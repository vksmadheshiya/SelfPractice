```python
import pandas as pd
```

```python
col1 = [1,2,3,4,5]
col2 = [45,56,67,34,34]
col3 = ["sudhansu", "kumar", "krish", "inueron", "one"]


df = pd.DataFrame({"col1": col1,
              "col2":col2,
              "col3":col3})
```

```python
def test(a):
  return a**2
```

```python
df['col4_4'] = df['col1'].apply(test)
```

```python
df
```

```python

def test1(a):
  if a=='s':
    return 's'
  else:
    return 'z'



df['col6'] = df['col3'].str[0].apply(test1)
df
```

```python
df['col6_6'] = df['col3'].str[0].apply(lambda a:'s' if a=='s' else 'z')
df
```
