Write the same shorting function without using short

```python
data_list = [65, 25, 12, 22, 11, 1, 2, 44, 3, 122, 23, 24]
new_list = []
while data_list:
  min = data_list[0]
  for x in data_list:
    if x>min:
      min = x
  new_list.append(min)
  data_list.remove(min)
print(new_list)
    
```

Fibbonacci series

```python
def F(n):
  if n==1:return 1
  elif n==0: return 0
  else: return F(n-1)+F(n-2)

for i in range(0, 15):
  print(F(i))
```

Check Whether a string is palindrome or not

```python
def Pali(palindrome):
  return True if palindrome[::-1].lower() == palindrome.lower() else False

print(Pali('Madam'))
```

```python
def isPalindrome(s):
  rev = "".join(reversed(s))
  if rev == s:
    return True
  return False

print(isPalindrome('madam'))
```

Impliment Binary Search

```python
def binary_search(array, target):
  lower = 0
  upper = len(array)

  while lower<upper:
    x = lower + (upper - lower) // 2

    val = array[x]
    if target == val:
      return x
    elif target > val:
      if lower == x:
        break;
    elif target < val:
      upper = x 
array = [1,435,64,6,46,4,52,2,5,4,52,3,5,]
array.sort()
print(binary_search(array, 5))

l = [65, 25, 12, 22, 11, 1, 2, 44, 3, 122, 23, 24]
l.sort()
# print(l)
binary_search(l, 11)
```

Delete Reoccurring Character

```python
def dlelteReOccurringCharacters(string):
  seenCharacter = set()
  outputString = ''
  for char in string:
    if char not in  seenCharacter:
      seenCharacter.add(char)
      outputString += char
  return outputString

```
