```python
# implement the logic to check if the number is a palindrome or not

n = int(input("Enter Number"))

temp = n
rev = 0

while(n>0):
  dig = n%10
  rev=rev*10 + dig
  n = n//10

if(temp==rev):
  print("The number is palindrome !")
else:
  print("The number is not a palindrome !")
```

```python
# print the pascal's pattern:

from math import factorial

n = 6

for x in range(n):
  for y in range(n-x+1):
    print(end=" ")

  for y in range(x+1):
    # formula for nCr = n!/((n-r)!*r!)
    print(factorial(x)//(factorial(y)*factorial(x-y)), end=' ') 
  print()
```
